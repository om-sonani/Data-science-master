### Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?

* The mean, median, and mode are measures of central tendency used in statistics to describe the typical or average value of a dataset.



1) The mean is the arithmetic average of a dataset, obtained by summing all the values in the dataset and dividing by the number of values. It is sensitive to outliers, meaning that extreme values can significantly impact the mean. The mean is commonly used when the dataset is normally distributed.



2) The median is the middle value of a dataset when the values are arranged in ascending or descending order. It is not affected by outliers and is used when the dataset has a skewed distribution or contains extreme values.



3) The mode is the most common value in a dataset. It is used when the dataset has a categorical or discrete numerical variable, and the most frequent value is of interest

### Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.

* Measures of dispersion such as range, variance, and standard deviation are used to describe the spread or variability of a dataset. These measures provide information about how spread out the values in the dataset are from the central tendency (mean, median, mode) of the data.

1) Range: The range is the difference between the maximum and minimum values in the dataset. It provides a simple measure of the spread of the data. However, it is sensitive to outliers in the dataset, as they can greatly affect the range.

2) Variance: The variance is the average of the squared differences from the mean of the dataset. It provides a measure of how much the values in the dataset deviate from the mean. A high variance indicates that the values are spread out more widely from the mean, while a low variance indicates that the values are closer to the mean.

3) Standard deviation: The standard deviation is the square root of the variance. It provides a measure of the spread of the data in the same units as the original data. A high standard deviation indicates that the values are more spread out from the mean, while a low standard deviation indicates that the values are closer to the mean.

### Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?

* In statistical analysis, covariance and correlation are commonly used to explore the relationship between two variables. Covariance is used to determine whether two variables tend to vary together, while correlation is used to determine the strength and direction of the relationship between two variables. Correlation is often preferred over covariance because it is standardized and provides a clearer understanding of the relationship between two variables.

### Q12. For a normal distribution data what is the relationship between its measure of central tendency?

* For a normal distribution data, the measures of central tendency, such as the mean, median, and mode, are all equal to each other.

> This is because the normal distribution is symmetric around its mean, which is also its median and mode. The mean represents the arithmetic average of the data, the median represents the middle value, and the mode represents the most frequently occurring value.

## Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and flexible web framework for building web applications using the Python programming language. It is classified as a micro-framework because it does not require particular tools or libraries. Flask allows developers to build web applications quickly and easily, without the overhead and complexity of a full-fledged web framework.

Some of the advantages of using Flask Framework include:

Easy to use: Flask is a simple and easy-to-use framework. It has a small core and comes with built-in development server and a debugger, which makes it easy to get started with and develop applications quickly.

Lightweight: Flask is a lightweight framework, which means that it does not require a lot of resources to run. This makes it an ideal choice for building small to medium-sized applications.

Flexible: Flask is highly flexible, which allows developers to create web applications according to their own requirements. Developers can use any database, templating engine, or other tools they want.

Modular: Flask is modular, which means that developers can add or remove features as needed. This makes it easy to scale an application, as developers can add new features when needed.

Large Community: Flask has a large and active community, which means that developers can find help and support when they need it. The community also develops and maintains many third-party libraries and extensions for Flask, which can be easily integrated into applications.

## Q3. What is App routing in Flask? Why do we use app routes?

In Flask, routing refers to the process of mapping URLs to the functions that handle them. Routing in Flask is accomplished using the @app.route() decorator.

>> There are several benefits of using app routes in Flask:

Readability: App routes in Flask make the code more readable and maintainable. They provide a clear and concise way to define the URLs that are available in an application.

Modularization: App routes help in modularizing the code. The application logic can be divided into different routes and functions, making it easy to organize and understand the code.

Flexibility: App routes provide flexibility to the developers to add or modify the functionality of the application. They can easily add new routes or modify existing ones without affecting the other parts of the application.

Testing: App routes make it easy to test the application. The developer can test each route and function separately, which helps in identifying the issues and debugging the code.

## Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
## route to show the following details:
## Company Name: ABC Corporation
## Location: India
## Contact Detail: 999-999-9999
## Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def details():
    return 'Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999'


if __name__=="__main__":
    app.run(host="0.0.0.0")
```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL for a given endpoint and any arguments passed to it. The endpoint is the name of the function that handles the route.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the index page'

@app.route('/user/<username>')
def profile(username):
    return f'This is the profile page for {username}'

with app.test_request_context():
    print(url_for('index')) # Output: /
    print(url_for('profile', username='John Doe')) # Output: /user/John%20Doe
```

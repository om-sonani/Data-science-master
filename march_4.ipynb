Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
titanic_dataset=sns.load_dataset("titanic")
```

```python
titanic_dataset
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic_dataset.age, y=titanic_dataset.fare, mode='markers'))
fig.show()
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px

# Load the "tips" dataset
tips = px.data.tips()

# Create a box plot using Plotly express
fig = px.box(tips, x="day", y="total_bill", color="sex")

# Show the plot
fig.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
import plotly.express as px

# Load the "tips" dataset
tips = px.data.tips()

# Create a histogram using Plotly express
fig = px.histogram(tips, x="sex", y="total_bill", color="day", pattern_shape="smoker")

# Show the plot
fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

```python
import plotly.express as px

# Load the "iris" dataset
iris = px.data.iris()

# Create a scatter matrix plot using Plotly express
fig = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species"
)

# Show the plot
fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

Distplot is a function in the seaborn library which is used to visualize the distribution of a single numerical variable in the form of a histogram with a kernel density estimate (KDE) plot overlayed on top.



```python
import plotly.figure_factory as ff
import numpy as np

x1 = np.random.randn(200) - 1
x2 = np.random.randn(200)
x3 = np.random.randn(200) + 1

hist_data = [x1, x2, x3]

group_labels = ['Group 1', 'Group 2', 'Group 3']
colors = ['#333F44', '#37AA9C', '#94F3E4']

# Create distplot with curve_type set to 'normal'
fig = ff.create_distplot(hist_data, group_labels, show_hist=False, colors=colors)

# Add title
fig.update_layout(title_text='Curve and Rug Plot')
fig.show()
```

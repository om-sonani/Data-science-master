>>Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

* Matplotlib is a Python data visualization library that allows you to create a wide range of static, animated, and interactive visualizations in Python. It is used to create high-quality graphics, charts, and plots in Python for data exploration and presentation. Matplotlib can be used in a wide variety of fields including data science, machine learning, statistics, finance, and engineering.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

1) Line plot: A line plot is used to display data points connected by straight lines. It is useful for showing trends in data over time.

2) Scatter plot: A scatter plot is used to show the relationship between two variables. It is useful for identifying trends, clusters, and outliers in data.

3) Bar plot: A bar plot is used to compare values across different categories. It is useful for visualizing the distribution of categorical data.

4) Histogram: A histogram is used to display the distribution of a numerical variable. It is useful for showing the frequency distribution of data.

5) Pie chart: A pie chart is used to show the proportion of different categories in a dataset. It is useful for showing the relative sizes of different categories in a dataset.

>>Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Data")
plt.show()
```

>>Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

* The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange plots in a grid format, where each plot is displayed in its own cell. This function is useful when you want to visualize multiple datasets or compare different visualizations side-by-side.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2)

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Adjust the layout of the subplots
plt.tight_layout()

# Display the figure
plt.show()
```

>>Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

* Bar plot: A bar plot is used to compare values across different categories. It is useful for visualizing the distribution of categorical data

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.show()
```

```python
plt.barh(company,profit)
plt.xlabel('profit')
plt.ylabel('company')
plt.show()
```

>>Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

* A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the median, quartiles, and outliers of the data in a compact way.

```python
#Quesiton 5
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()

ax.boxplot(data)

# Add labels and title
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box Plot of Two Datasets')

plt.show()
```

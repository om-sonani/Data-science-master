```python
def validate_password(password):
    import re
    
    if len(password) != 10:
        return "Invalid Password"
    
    if not re.search("[a-z]{2,}", password):
        return "Invalid Password"
    
    if not re.search("[A-Z]{2,}", password):
        return "Invalid Password"
    
    if not re.search("[0-9]", password):
        return "Invalid Password"
    
    if not re.search("[!@#$%^&*]{3,}", password):
        return "Invalid Password"
    
    return "Valid Password"

validate_password("omOM@#!123")
```

```python
def starts_with(string_list, letter):
    return [x for x in string_list if x.startswith(letter)]
```

```python
starts_with("omsonani","o")
```

```python
#using list comprehension

def only_numeric(string_list):
    return [x for x in string_list if x.isdigit()]

only_numeric("123")

```

```python
fruits = [("mango", 99), ("orange", 80), ("grapes", 1000)]
sorted_fruits = sorted(fruits, key=lambda x: x[1])
sorted_fruits

```
